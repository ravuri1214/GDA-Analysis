###   3) Whats the world GDP in 1969?

```python
df_pr = df[df['Year'] == 1969]
count = 0
for i in df_pr.values:
    count += i[3]
print(count)
    
        
```

### 4) Whats the % of GDP growth of India in Year 1994 (*as compare to 1993)?

```python
df_pr1= df[df['Country Name'] == 'India'] 
count = 0
count = 1
for i in df_pr1.values:
    if i[2] == 1993:
        count = i[3]
    if i[2] == 1994: 
        count1 = i[3]
print(((count1 - count) / count)*100)
```

```python
df_pr = df[df['Country Name'] == 'India']
data = df_pr.values
GDP_Values = [0]
for i in range(1, len(data)):
    curr = data[i][3]
    prev = data[i-1][3]
    values = round(((curr - prev) / prev)*100,2)
    GDP_Values.append((values))
df_pr = df_pr.assign(GDP = GDP_Values)   
    
```
